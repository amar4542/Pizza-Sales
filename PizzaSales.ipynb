```python
import pandas as pd 
```

```python
df = pd.read_csv('pizza_sales.csv')
df.head()
```

```python
df.drop_duplicates()
df
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format ='%d-%m-%Y')
```

```python
df['order_time'] = pd.to_datetime(df['order_time'],format = '%H:%M:%S').dt.time
```

```python
df.info()
```

```python
df.describe
```

```python
import matplotlib.pyplot as plt 
```

```python
sales_by_size =df.groupby('pizza_size')['total_price'].sum()
plt.figure(figsize=(5,3))
plt.bar(sales_by_size.index, sales_by_size.values, color = 'Green',width = 0.5)
plt.xlabel("Pizza Size")
plt.ylabel("Total Sales")
plt.title("Total Sales by Pizza Size")
#plt.grid()
plt.show()
```

```python
Category_sales = df.groupby('pizza_category')['total_price'].sum()

plt.figure(figsize=(4,4))
colors = ['orange','lightblue','Red','yellow']
plt.pie(Category_sales, labels=Category_sales.index, autopct='%1.2f%%',colors = colors,explode = [0.1,0.1,0.1,0.1])
plt.axis('equal')
plt.title("Total Sales by Pizza Category")
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')
df['order_date'].dt.year.unique()
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['total_price'].sum()

plt.figure(figsize=(7,4))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o',color = 'purple')
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.grid()
plt.show()
```

```python
pizza_sales = df.groupby('pizza_name')['quantity'].sum().sort_values()
plt.figure(figsize=(7,6))
plt.barh(pizza_sales.index, pizza_sales.values, color='orange')
plt.xlabel("Total Quantity Sold")
plt.ylabel("Pizza Name")
plt.title("Total Sales by Pizza Name")
plt.show()
```
